```python
housing_imp = housing_dirty.copy()
replace_with = housing_imp.mean(numeric_only=True)
housing_imp = housing_imp.fillna(replace_with)
housing_imp.head()
```

## 10 - Perform standardization

```python
features = ['ZN', 'INDUS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'PRICE']
```

```python
housing_std = housing_dirty.copy()
to_std = housing_std[features]

housing_std[features] = (to_std - to_std.mean()) / to_std.std()
housing_std.head(10)
```
